### A5 and effect of measures (RfR, Evacuation)

More RfR costs should mean more projects carried out and more evacuation costs probably earlier warning.

```python
interest_A5 = results[['A.5_Expected Total Damage', 'A.5_Total Dike Investment Costs', 'A.5_Total Expected Number of Deaths',
                      'RfR Total Costs','Total Expected Evacuation Costs']]
interest_A5
```

```python
sns.pairplot(interest_A5)
plt.show()
```

No clear correlation between RfR and/or dike investment costs and the damages / deaths in Deventer. Expected number of deaths and damage cost somehow related. The same goes for expected total damage and expected evacuation costs. 

### Us vs dike failure

```python
input_pfail_vs_output_A5 = results[['A.5_Expected Total Damage', 'A.5_Total Dike Investment Costs', 'A.5_Total Expected Number of Deaths',
                             'A.1_pfail','A.2_pfail', 'A.3_pfail','A.4_pfail', 'A.5_pfail']]
```

```python
sns.pairplot(input_pfail_vs_output_A5)
plt.savefig('figures/dike_failure_open_exploration.png')
plt.show()
```

Interesting to note: A.5 Total Dike investment costs does not influence dike failure in any visible way. However, the lower, the more damage can occur, and after ca. 1.5 1e9 costs no too high casualties can be expected in Deventer. 
Otherwise interesting: dike failure in other dike rings does not seem to affect our damage and human cost. Only failure of A5 has an effect: but surprisingly does a higher dike failure probability not result in more damage or deaths.
