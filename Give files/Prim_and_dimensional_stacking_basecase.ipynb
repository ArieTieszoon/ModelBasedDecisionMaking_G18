# Prim and dimensional stacking on base case

```python
import pandas as pd
from ema_workbench.analysis import prim
import matplotlib.pyplot as plt
from ema_workbench.analysis import dimensional_stacking
```

```python
experiments = pd.read_csv('Outcomes/Experiments_from_Exploration_10000scenarios_pf5.csv')
outcomes_total_dikerings_load_DF = pd.read_csv('Outcomes/Proccesed_basecase_fully_aggregated_10000scenarios_pf5.csv')
```

## Deaths

In this section analysis are performed on the simulated number of deaths by the dike_function.py model

### Total deaths in dikering 5

```python
data = outcomes_total_dikerings_load_DF['A.5_Total Expected Number of Deaths']

#desirable risk by Dutch government (1/100000) times citizens of Deventer (+/- 100000 in jan 2019)
y = data < 1

prim_alg = prim.Prim(experiments,y, threshold=0.8)

box1 = prim_alg.find_box()
```

#### Treshold of Dutch government

```python
fig = box1.show_tradeoff()
fig.suptitle('Dikering 5 threshold of deaths tradeoff', fontsize=16)
plt.savefig('Figures_dewy/dikering5_deaths_tradeoff.png')
plt.show()
```

```python
box1.inspect(0)
```

```python
fig = box1.inspect(1,style='graph')
fig.suptitle('Dikering 5 threshold of deaths constraints', fontsize=16)
plt.savefig('Figures_dewy/dikering5_deaths_constrains.png')
plt.show()
```

```python
box1.select(1)
```

```python
fig = box1.show_pairs_scatter()
```

Analysis: the outcomes of the model suggest that in dikering 5 causialties will remain under the treshold of the  Dutch goverment in the observed scenarios when the pfail fo the dike is 0.05 or higher.

#### Dimensinal stacking

```python
dimensional_stacking.create_pivot_plot(experiments, y)
plt.show()
```

Analysis: the dimensional stacking shows as well that most important indicator for the outcomes that go over the treshold is the 'A5_pfail'

In this case, when Deventer has a low "pfail" it will only experience damage when upstreamd dikes do not break: so "pfail" of the dikes is sufficient.

## Conclusion

The scenario discovery showed that outcomes with a number of deaths exceeding the treshold appear to be linked at the "pfail" rates of dikering 1 and 3. For dikering 4 and 5, these outcomes appear to be linked only at their own "pfail" rate, however, in almost all cases the treshold is met.

In terms of damage intersting behaviour is found. Outcomes that result in damage for dikerings 4 and 5 do not only relate with a low "pfail" of the own dikering, but also with minimum "pfail" rates of upstream dikerings. This dependency confirms the complex character of this project.
In the whole IJssel delta the scenario's with the highest damage are related to a low "pfail" in dikering 1 and low discount rates. Outcomes with low damage are related with minimum treshold of "pfails" for all dikerings, where dikering 1 and 3 have relative high numbers and dikerig 4 low threshold.
